```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Marketing_Design_Dataset.csv")
```

```python
df.head(4)
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Campaign_Budget'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Campaign Budget ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Campaign Budget')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(range(len(df)), df['Campaign_Budget'], color='skyblue', label='Campaign Budget')
plt.xlabel('Campaign Index')
plt.ylabel('Campaign Budget ($)')
plt.title('Campaign Budget Across Different Campaigns')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Ad_Click_Rate'], df['Conversion_Rate'], alpha=0.5, color='purple')
plt.xlabel('Ad Click Rate (%)')
plt.ylabel('Conversion Rate (%)')
plt.title('Ad Click Rate vs Conversion Rate')
plt.show()
```
